```python
import pandas as pd
import numpy as np

data=pd.read_csv('I://تطبيق machine//4_5901974192206645743.csv')
```

```python
data=data.drop(['id','age','ba','pot','su','pc','pcv','wc','rc','cad','dm','ane','bp','rbc','pcc','bgr','appet','pe'],axis=1)
for i ,j in enumerate(data):
    if data[j].dtypes=='O':
        data[j].str.strip()

print(data.nunique())
```

```python
# data=data.apply(lambda x: x.value_counts.index[0])
for i  in  (data.columns):
    if data[i].isnull:
        data[i]=data[i].fillna(data[i].mode()[0])
data.head(10)
```

```python
data['classification'].replace('ckd\t','ckd',inplace=True)
```

```python
data.nunique()
```

```python
data['htn']=data['htn'].map({'yes':1,'no':0})

data['classification']=data['classification'].map({'ckd':1,'notckd':0})
# data.head(20)
```

```python
data.head(20)
```

```python
x=data.drop(['classification'],axis=1)
y=data['classification']

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
x.columns
```

```python
# create the model
from sklearn.linear_model import LogisticRegression
model= LogisticRegression()
model.fit(x_train,y_train)

pred=model.predict(x_test)
print(model.score(x_test,y_test))
```

```python
import pickle
file_name='model_ckd.sav'
pickle.dump(model, open(file_name,'wb'))
```

```python
# loading the model saved
loaded_model= pickle.load(open('model_ckd.sav', 'rb'))
loaded_model
```
